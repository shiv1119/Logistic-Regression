```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ChurnData.csv")
```

```python
df.head(10)
```

```python
df = df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
df['churn'] = df['churn'].astype('int')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
x = np.asanyarray(df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']])
x[0:10]
```

```python
y = np.asanyarray(df['churn'])
y[0:10]
```

```python
from sklearn import preprocessing
x = preprocessing.StandardScaler().fit(x).transform(x)
x[0:10]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 4)
print("Shape of train set : ", x_train.shape, y_train.shape)
print("Shape of test set : ", x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=0.01, solver='liblinear').fit(x_train, y_train)
```

```python
lr
```

```python
predictions = lr.predict(x_test)
```

```python
predictions
```

```python
predictions_prob = lr.predict_proba(x_test)
```

```python
predictions_prob
```

```python
from sklearn import metrics
j_score = metrics.jaccard_score(y_test, predictions, pos_label=0)
print("Jaccard Score : {0:.5f}".format(j_score))
```

```python
print (metrics.classification_report(y_test, predictions))
```

```python
metrics.log_loss(y_test, predictions_prob)
```
